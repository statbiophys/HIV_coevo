```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import math
```

```python
def read_params(path):
    """Read a parameter file"""
    params = dict()
    f = open(path, "r")
    for l in f.readlines():
        try:
            params[l.split('\t')[0]] = float(l.split('\t')[1])
        except ValueError:
            if ',' not in l.split()[1]:
                params[l.split('\t')[0]] = l.split('\t')[1]
            else:
                params[l.split('\t')[0]] = np.array(l.split('\t')[1].split(',')[:], dtype=float)
    f.close()
    return params


def read_2dtraj(path):
    f = open(path, 'r')
    traj = []
    for line in f.readlines():
        traj.append(line.split('\t')[:-1])
    f.close()
    return np.array(traj, dtype=float)


def smooth_traj(times, traj, window):
    smooth_tr, smooth_times = [], []
    for t in range(len(times)-window):
        smooth_tr.append(traj[t:t+window].mean())
        smooth_times.append(times[t:t+window].mean())
    return smooth_times, smooth_tr

def sel(c, beta):
    return c * beta / 50

def build_rho_dict(shift=0):
    
    rhos_d = dict()
    for pair in product(range(nc1), range(ndt)):
        fi, dti = pair[0], pair[1]
        dt = int(params['turn_long_dt'][dti])
        
        # This traj has all the turnovers of all the realizations one after the other
        traj = read_2dtraj(data_dir+'turn'+str(fi)+'_dt'+str(dt)+'.txt')
        # Creating a trajectory for each realization (first dim is realization)
        traj = np.reshape(traj, (r, int(len(traj)/r), 2)) 
        
        rho_list = []
        l = len(traj[0][:,0])
        for tr in traj:
            if shift >= 0:
                rho = spearmanr(tr[:l-shift,0], tr[shift:,1])[0]
            else:
                rho = spearmanr(tr[-shift:,0], tr[:l+shift,1])[0]
            if not math.isnan(rho):
                rho_list.append(rho)
            
        rhos_d[pair] = rho_list
        print(pair)
        
    return rhos_d

def compute_null_corr_std(n_points, R=3000):
    rhos = []
    for _ in range(R):
        r,p = spearmanr(np.random.rand(n_points), np.random.rand(n_points))
        rhos.append(r)
    return np.std(rhos)
```

```python
data_dir = 'scan3/'
params = read_params(data_dir+'param.txt')
params
```

```python
from scipy.stats import spearmanr
from itertools import product

r = int(params['n_realizations'])
nc1, ndt = len(params['sel_coefs1']), len(params['turn_long_dt'])
```

```python
shift = 0
rho_d = build_rho_dict(shift)

n_points = int(params['n_turn_points'])
std_null = dict()
std_null[n_points] = compute_null_corr_std(n_points)
    
x = np.linspace(-1,1,100)
```

```python
means, stds, counts = np.zeros((nc1, ndt)),np.zeros((nc1, ndt)),np.zeros((nc1, ndt))
counts_th = 100
for f1i in range(nc1):
    for dti in range(len(params['turn_long_dt'])):
        counts[f1i,dti] = len(rho_d[(f1i,dti)])
        if counts[f1i,dti] > counts_th:
            means[f1i,dti] = np.mean(rho_d[(f1i,dti)])/std_null[n_points]
            stds[f1i,dti] = np.std(rho_d[(f1i,dti)])/std_null[n_points]
        
        
from matplotlib import cm

cmap = cm.get_cmap('viridis')
def normalize(val, vmin, vmax):
    return (val-vmin)/(vmax-vmin)
```

```python
fig, (ax1, ax2) = plt.subplots(1,2,figsize=(12,6))
ax1.set_title('L=1, N=%i, $\mu$=%.6g' %(params['pop_size1'], params['mut_rate1']), fontsize=14)
ax1.set_xlabel(r'Turnover delta time, $\Delta t$', fontsize=14)
ax2.set_xlabel(r'Rescaled turnover delta time, $\frac{2 \Delta t}{T_{e1}+T_{e2}}$', fontsize=14)
ax1.set_ylabel('Average rescaled corr coef', fontsize=14)
ax1.set_xscale('log')
ax2.set_xscale('log')
for f1i in range(nc1):
    s1, s2 = params['sel_coefs1'][f1i], params['sel_coefs2'][f1i]
    N1, N2 = params['pop_size1'], params['pop_size2']
    m1, m2 = params['mut_rate1'], params['mut_rate2']
    Te1, Te2 = 1/N1/m1/abs(s1), 1/N2/m2/abs(s2)
    Ts1, Ts2 = np.log(N1*abs(s1))/abs(s1), np.log(N2*abs(s2))/abs(s2)
    av_Te = (Te1 + Te2)/2.0
    ratio_min = min(Te1/Ts1, Te1/Ts2, Te2/Ts1, Te2/Ts2)

    #if ratio_min < TeTs_th:
    #    continue
        
    ax1.plot(params['turn_long_dt'], means[f1i, :],
             label=r's=%.3g, %.3g, $T_e$/$T_s$=%.2g'%(s1,s2,ratio_min), lw=2)
    ax2.plot(params['turn_long_dt']/av_Te, means[f1i, :], lw=2)
    
ax1.legend()
plt.tight_layout()
plt.savefig(data_dir+'av_corr_te.svg', dpi=200)
```

### Generating average turnover correlations 

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import seaborn as sns
import math

seq_len = 50
```

```python
def read_params(path):
    """Read a parameter file"""
    params = dict()
    f = open(path, "r")
    for l in f.readlines():
        try:
            params[l.split('\t')[0]] = float(l.split('\t')[1])
        except ValueError:
            if ',' not in l.split()[1]:
                params[l.split('\t')[0]] = l.split('\t')[1]
            else:
                params[l.split('\t')[0]] = np.array(l.split('\t')[1].split(',')[:], dtype=float)
    f.close()
    return params


def read_2dtraj(path):
    f = open(path, 'r')
    traj = []
    for line in f.readlines():
        traj.append(line.split('\t')[:-1])
    f.close()
    return np.array(traj, dtype=float)


def smooth_traj(times, traj, window):
    smooth_tr, smooth_times = [], []
    for t in range(len(times)-window):
        smooth_tr.append(traj[t:t+window].mean())
        smooth_times.append(times[t:t+window].mean())
    return smooth_times, smooth_tr

def build_rho_dict(shift=0):
    
    rhos_d = dict()
    for tripl in product(range(ns), range(ns), range(ndt)):
        fi1, fi2, dti = tripl[0], tripl[1], tripl[2]
        dt = int(params['turn_long_dt'][dti])
        
        # This traj has all the turnovers of all the realizations one after the other
        traj = read_2dtraj(data_dir+'turn'+str(fi1)+'_'+str(fi2)+'_dt'+str(dt)+'.txt')
        # Creating a trajectory for each realization (first dim is realization)
        traj = np.reshape(traj, (r, int(len(traj)/r), 2)) 
        
        rho_list = []
        l = len(traj[0][:,0])
        for tr in traj:
            if shift >= 0:
                rho = spearmanr(tr[:l-shift,0], tr[shift:,1])[0]
            else:
                rho = spearmanr(tr[-shift:,0], tr[:l+shift,1])[0]
            if not math.isnan(rho):
                rho_list.append(rho)
            
        rhos_d[tripl] = rho_list
        print(tripl)
        
    return rhos_d

def compute_null_corr_std(n_points, R=3000):
    rhos = []
    for _ in range(R):
        r,p = spearmanr(np.random.rand(n_points), np.random.rand(n_points))
        rhos.append(r)
    return np.std(rhos)
```

```python
data_dir = 'scan_s1/'
params = read_params(data_dir+'param.txt')
params
```

```python
from scipy.stats import spearmanr
from itertools import product

r = int(params['n_realizations'])
ns, ndt = len(params['sel_coefs']), len(params['turn_long_dt'])
```

```python
shift = 0
rho_d = build_rho_dict(shift)

n_points = int(params['n_turn_points'])
std_null = dict()
std_null[n_points] = compute_null_corr_std(n_points)
    
x = np.linspace(-1,1,100)
```

```python
means, stds, counts = np.zeros((ns, ns, ndt)),np.zeros((ns, ns, ndt)),np.zeros((ns, ns, ndt))
counts_th = 10
for f1i, f2i, dti in product(range(ns), range(ns), range(ndt)):
    counts[f1i,f2i,dti] = len(rho_d[(f1i,f2i,dti)])
    if counts[f1i,f2i,dti] > counts_th:
        means[f1i,f2i,dti] = np.mean(rho_d[(f1i,f2i,dti)])/std_null[n_points]
        stds[f1i,f2i,dti] = np.std(rho_d[(f1i,f2i,dti)])/std_null[n_points]
        
        
from matplotlib import cm

cmap = cm.get_cmap('viridis')
def normalize(val, vmin, vmax):
    return (val-vmin)/(vmax-vmin)
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1,3,figsize=(12,3.5))
max_abs_corr = np.max([abs(np.min(means)), abs(np.max(means))])

dt_i = 0
dt = params['turn_long_dt'][dt_i]
ax1.set_title(r'$\Delta t$ = '+str(dt))
sns.heatmap(means[:,:,dt_i], ax=ax1, 
            vmin=max_abs_corr, vmax=-max_abs_corr, cmap='bwr', 
            xticklabels=params['sel_coefs'], yticklabels=params['sel_coefs'])
ax1.invert_yaxis()

dt_i = 1
dt = params['turn_long_dt'][dt_i]
ax2.set_title(r'$\Delta t$ = '+str(dt))
sns.heatmap(means[:,:,dt_i], ax=ax2, 
            vmin=max_abs_corr, vmax=-max_abs_corr, cmap='bwr', 
            xticklabels=params['sel_coefs'], yticklabels=False)
ax2.invert_yaxis()

dt_i = 2
dt = params['turn_long_dt'][dt_i]
ax3.set_title(r'$\Delta t$ = '+str(dt))
sns.heatmap(means[:,:,dt_i], ax=ax3, 
            vmin=max_abs_corr, vmax=-max_abs_corr, cmap='bwr', 
            xticklabels=params['sel_coefs'], yticklabels=False)
ax3.invert_yaxis()

plt.savefig(data_dir+'/turn_heatmap.svg')
```
